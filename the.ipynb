```python
import numpy as np
from matplotlib import pyplot as plt

def l_j(r, sigma=1.0, epsilon=1.0):
    """Lennard-Jones potential."""
    return 4 * epsilon * ((sigma / r) ** 12 - (sigma / r) ** 6)

def neg(r, sigma=1.0, epsilon=1.0):
    """Attractive part of the Lennard-Jones potential."""
    return 4 * epsilon * (- (sigma / r) ** 6)

def l_j_force(r, sigma=1.0, epsilon=1.0):
    """Lennard-Jones force."""
    return 24 * epsilon * (2 * (sigma / r) ** 12 - (sigma / r) ** 6) / r
```

```python
rs = np.linspace(0.9, 1.5, 100)
# plt.figure(figsize=(10, 20))
plt.plot(rs, l_j(rs), label='Lennard-Jones potential')
plt.plot(rs, neg(rs), label='Attractive part of Lennard-Jones potential')
```
